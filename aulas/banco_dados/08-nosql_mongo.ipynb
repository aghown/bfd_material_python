### Importando o Driver

````python
import pymongo
````

### Criando uma conexão

````python
# localhost
conn = pymongo.MongoClient("mongodb://localhost:27017/")
````

ou

````python
# servidor mongo atlas
conn = pymongo.MongoClient("mongodb+srv://fred:senha@cluster0.wu6w6nh.mongodb.net/?appName=Cluster0")
````

### Criando um banco

````python
banco = conn["loja4"]
````

### Criando uma coleção

````python
colecao_users = banco["users"]
````

### Adicionando um documento

````python
colecao_users.insert_one({"nome":"Fred", "idade":25, "endereco":{"cep":82738273, "logradouro": "Rua do seu zé", "numero": 34, "Referencia": None}})
````

````python
colecao_users.insert_one({"nome":"Fredão", "idade":25, "endereco":{"cep":82738273, "logradouro": "Rua do seu zé", "numero": 34, "Referencia": None},"profissão":"Programador"})
````

### **STOP!!!**
Hora de falar de variáveis de ambiente

Precisa instalar a biblioteca **python-dotenv**

````python
python -m pip install python-dotenv
````

Criar o arquivo que irá conter as variáveis de ambientes.

A convenção é usar o nome **.env** e seu conteúdo deve seguir um padrão chave valor como o seguinte:
```
SECRET_KEY=mysecretkey
TOKEN=mytoken
DEBUG=True
```


No caso dessa aula, vamos armazenar os dados para conexão com o banco no servidor do mongodb atlas:

No arquivo agora temos que importar as bibliotecas os e dotenv: 

````python
import os
from dotenv import load_dotenv
````

e agora, puxar os dados do arquivo .env:

````python

# Carregando as variáveis de ambiente do arquivo .env
load_dotenv()
# Acessando a variável de ambiente API_KEY
bancotoken = os.getenv("MONGODB")
print(bancotoken)

conn = pymongo.MongoClient(bancotoken)
````

### RETORNANDO para o conteúdo de mongoDB

### Realizar consultas

````python
users = colecao_users.find()
print(users)
for user in users:
    print(user)
conn.close()
````
